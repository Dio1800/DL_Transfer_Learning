# Detect human faces

Alternative python code to detect the region(s) of a image with human face(s). 

OBS: Download files:
 OpensCV - https://github.com/opencv/opencv/blob/4.x/data/haarcascades/haarcascade_frontalface_default.xml
 image - http://images.cocodataset.org/val2017/000000541055.jpg
 image2 - http://images.cocodataset.org/val2017/000000131444.jpg

## Import libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Load image

```python
face_classifier = cv2.CascadeClassifier('OpenCV_model\haarcascade_frontalface_default.xml')

image = cv2.imread('OpenCV_model/000000131444.jpg')
plt.imshow(image)
plt.show()

image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
plt.show()
```

```python
# run detection model
faces = face_classifier.detectMultiScale(image_gray, 1.3, 5)

for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)

plt.imshow(image)
plt.show()

#cv2.imshow('image',image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
