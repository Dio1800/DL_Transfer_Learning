```python
# pip install opencv-python
import cv2
import numpy as np

img = cv2.imread('cat_16.jpg')
cv2.imshow("image - original", img)
img_2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("image - Gray", img_2)
suave = cv2.GaussianBlur(img_2, (7, 7), 0) #Apply blur
(T,bin) = cv2.threshold(suave, 160,255, cv2.THRESH_BINARY)
(T,binI) = cv2.threshold(suave, 160,255, cv2.THRESH_BINARY_INV)
resultado = np.vstack([
    np.hstack([suave,bin]),
    np.hstack([binI, cv2.bitwise_and(img_2, img_2, mask = binI)])
    ])
cv2.imshow("image black and white", resultado)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
plt.figure()

#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(3,2) 

# use the created array to output your multiple images. In this case I have stacked 6 images in 3 rows and 2 columns
axarr[0][0].imshow(img)
axarr[0][1].imshow(img_2)
axarr[1][0].imshow(suave)
axarr[1][1].imshow(bin)
axarr[2][0].imshow(binI)
axarr[2][1].imshow(cv2.bitwise_and(img_2, img_2, mask = binI))
```
